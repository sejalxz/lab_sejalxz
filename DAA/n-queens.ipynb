# Assignment No 5

Design n-Queens matrix having first Queen placed. Use backtracking to place remaining Queens to generate the final n-queenâ€˜s matrix.

```python
class Solution:
    def solveNQueens(self, n):
        
        def populateBoard(board, n):
            t_ans = []
            for i in range(len(board)):
                t = ""
                for j in range(n):
                    t += board[i][j]
                t_ans.append(t)
            return t_ans
        
        def issafe(r,c):
            for i in range(n):
                if board[i][c] == 'Q':
                    return False
                if r - i >= 0 and c - i >= 0 and board[r-i][c-i] == 'Q':
                    return False
                if r - i >= 0 and c + i < n and board[r-i][c+i] == 'Q':
                    return False
            return True
        
        def isValid(r,c):
            for row in range(0,n):
                if(board[row][c] == 'Q'): return False
                
            for col in range(0,n):
                if(board[r][col] == 'Q'): return False
                
            #upper left
            row = r
            col = c
            
            while (row >= 0 and col >= 0):
                if(board[row][col] == 'Q'): return False
                row = row - 1
                col = col - 1
                
            #upper right
            row = r
            col = c
            
            while (row >= 0 and col < n):
                if(board[row][col] == 'Q'): return False
                row = row - 1
                col = col + 1
                
            #lower left
            row = r
            col = c
            
            while (row < n and col >= 0):
                if(board[row][col] == 'Q'): return False
                row = row + 1
                col = col - 1
                
            #lower right
            row = r
            col = c
            
            while (row < n and col < n):
                if(board[row][col] == 'Q'): return False
                row = row + 1
                col = col + 1
                
            return True
                
        def solve(r):
            n = len(board)
            if r == n:
#                 print(board)
#                 ans.append(["".join(i) for i in board])
                t=populateBoard(board,n)
                ans.append(t)
                return 
            for c in range(0,n):
                if isValid(r,c):
                    board[r][c] = 'Q'
                    solve(r+1)
                    board[r][c] = '.'
                    
        def printAns(ans,n):
            print("No of solutions : ", len(ans),"\n")
            for i in range (len(ans)):
                for j in range(n):
                    print(ans[i][j])
                print()
                    
        board = [['.']*n for i in range(n)]
#         board[1][1] = 'Q'
        ans =[]
        solve(0) 
        printAns(ans,n)
    
s = Solution()
ans = s.solveNQueens(4)
```
